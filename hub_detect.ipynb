```python
import torch
from IPython.display import Image, display
import os

# 加载本地 YOLOv5s 模型
model = torch.hub.load('./', 'yolov5s', source='local')

# 测试图片路径
img_path = "./data/images/zidane.jpg"

# 推理
results = model(img_path)

# 创建保存目录
save_dir = "runs/detect/exp_test"
os.makedirs(save_dir, exist_ok=True)

# 保存结果图片到指定目录
results.save(save_dir)

# 获取保存的图片路径（YOLOv5默认只保存第一张图片到 files）
for f in results.files:
    print("Saved image:", f)
    display(Image(f))
```

```python
import torch
import cv2
from IPython.display import Image, display

# 加载本地 YOLOv5 模型
model = torch.hub.load('./', "custom","yolov5x", source="local")

# 使用 GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

# 图片路径
img_path = "./data/images/zidane.jpg"

# 读取图像并确保可写
img = cv2.imread(img_path)
img = img.copy()  # <--- 关键，创建可写副本

# 推理
results = model(img)

# 保存结果
results.save(save_dir="./runs/detect/")

# Notebook 显示结果
display(Image(filename='./runs/detect/exp/zidane.jpg'))

```

```python
import torch
import cv2

# 加载本地 YOLOv5 模型
model = torch.hub.load('./', "custom","runs/train/exp3/weights/best.pt", source="local")

# 使用 GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

# 图片路径
img_path = "./dataset/images/train/51.jpg"

# 读取图像并确保可写
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # 转成 RGB

#img = img.copy()  # <--- 关键，创建可写副本

# 推理
results = model(img)

results.show()
```
